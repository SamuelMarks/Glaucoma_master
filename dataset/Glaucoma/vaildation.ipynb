```python
from __future__ import print_function
import pandas as pd
import sys
import os
from config import PROJECT
sys.path = ["C:\\Users\\infinigru\\Anaconda3\\envs\\prac\\lib\\site-packages"] + sys.path
import cv2
import matplotlib.pyplot as plt
```

```python
v_label = pd.read_csv('vaildation_label.csv')
```

```python
location = os.path.join(PROJECT, 'dataset', 'REFUGE-Validation400')
```

```python
save_location = os.path.join(PROJECT, 'dataset', 'Glaucoma', 'Test')
```

```python
img = None
for _,rows in v_label.iterrows():
    img = cv2.imread(os.path.join(location,rows['ImgName']), cv2.IMREAD_COLOR)
    if rows['Glaucoma Label'] == 0:
        imname = os.path.join(save_location,'Non-Glaucoma',rows['ImgName'])
        cv2.imwrite(imname,img)
    else:
        print(rows['Glaucoma Label'],'| ',rows['ImgName']) 
        imname = os.path.join(save_location,'Glaucoma',rows['ImgName'])
        cv2.imwrite(imname,img)
```
